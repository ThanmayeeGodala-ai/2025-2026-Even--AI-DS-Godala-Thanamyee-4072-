```python
#13EDAoperations
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data=[10,12,15,18,20,22,25,30,35,40]
df=pd.DataFrame(data,columns=["values"])
print("Dataset:")
print(df)
print("\n summary statistics: ")
print(df.describe())
print("\n variance:",np.var(data))
print("\nstandard deviation:",np.std(data))
plt.hist(data)
plt.xlabel("values")
plt.ylabel("frequency")
plt.title("Data distribution")
plt.show()
mean_value=np.mean(data)
median_value=np.median(data)
print("\n inference")
print("mean value:",mean_value)
print("median value",median_value)
if mean_value>median_value:
  print("data is slightly right_skewed")
elif mean_value<median_value:
  print("data is slightly left_skewed")
else:
  print("data is symmetric")
```

```python
#12 datasetoperations
import pandas as pd
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)
print("Original dataset:")
print(df.head())
print("Records where total bill>20 and tip>3:")
filtered_and=df[(df['total_bill']>20) & (df['tip']>3)]
print(filtered_and)
print("Records where day is saturday or sunday")
filtered_or=df[(df['day']=='Sat') | (df['day']=='Sun')]
print(filtered_or)
print("Records where smoker is NOT yes:")
filtered_not=df[~(df['smoker']=='Yes')]
print(filtered_not)
print("\nFemale and non smoker and total bill>15")
combined = df[(df['sex'] == 'Female') & (df['smoker'] == 'No') & (df['total_bill'] > 15)]
print(combined)
print("\nNumber of records satisfying combined condition:",len(combined))
print("Sorted by total bill(Ascending):")
sorted_df=df.sort_values(by='total_bill')
print(sorted_df.head())
print("Sorted by Tip(Descending):")
sorted=df.sort_values(by='tip',ascending=False)
print(sorted.head())
print("\nSorted by day and total bill")
sorted_multi=df.sort_values(by=['day','total_bill'])
print(sorted_multi.head())
print("\nAverage total bill by day")
avg_bill=df.groupby('day')['total_bill'].mean()
print(avg_bill)
print("\nAverage tip by sex and smoker")
avg_tip=df.groupby(['sex','smoker'])['tip'].mean()
print(avg_tip)
print("\ncount of records by day")
count_day=df['day'].value_counts()
print(count_day)
```

```python
#14EDA(dataset url)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
url="https://raw.githubusercontent.com/dataprofessor/data/master/BostonHousing.csv"
df=pd.read_csv(url)
print("dataset(first 10 rows):")
print(df.head(10))
data=df['rm']
print("\n summary statistics:")
print(data.describe())
print("\n variance:",np.var(data))
print("\n standard deviation:",np.std(data))
plt.hist(data,bins=10,color="skyblue",edgecolor="black")
plt.axvline(np.mean(data),color="red",linestyle="dashed",linewidth=7,label=f"mean={np.mean(data):.2f}")
plt.axvline(np.median(data),color="green",linestyle="dashed",linewidth=2,label=f"meadian={np.median(data):.2f}")
plt.xlabel("rm")
plt.ylabel("frequency")
plt.title("distribution of sepal length")
plt.legend()
plt.show()
mean_value=np.mean(data)
meadian_value=np.median(data)
print("mean:",mean_value)
print("median:",meadian_value)
if mean_value>median_value:
  print("data is slightly right_skewed")
elif mean_value<median_value:
  print("data is slightly left_skewed")
else:
  print("data is symmetric")
```

```python
# Import required libraries
#15 data visualization
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read dataset from URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

# Step 2: Display first 5 rows
print("First 5 rows of dataset:")
print(df.head())

# Step 3: Dataset information
print("\nDataset Info:")
print(df.info())

# -----------------------------
# Data Visualization
# -----------------------------

#Line Plot - Sepal Length
plt.figure()
plt.plot(df['sepal_length'])
plt.title("Sepal Length Line Plot")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.show()


#Histogram - Petal Length
plt.figure()
plt.hist(df['petal_length'], bins=20)
plt.title("Petal Length Distribution")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()


#Scatter Plot - Sepal vs Petal Length
plt.figure()
plt.scatter(df['sepal_length'], df['petal_length'])
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()


#Bar Plot - Species Count
plt.figure()
df['species'].value_counts().plot(kind='bar')
plt.title("Species Count")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()


#Pie Chart - Species Distribution
plt.figure()
species_count = df['species'].value_counts()

plt.pie(species_count,
        labels=species_count.index,
        autopct='%1.1f%%',
        startangle=90)

plt.title("Species Distribution Pie Chart")
plt.axis('equal')  # Makes it a perfect circle
plt.show()
```
